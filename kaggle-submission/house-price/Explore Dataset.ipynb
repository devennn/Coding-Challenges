```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
```

```python
df = pd.read_csv('train.csv')
df.head(100)
```

```python
df.tail(100)
```

```python
del df['Alley']
del df['PoolQC']
del df['Fence']
del df['MiscFeature']

df
```

```python
sub_class = []
for i in range(len(df['MSSubClass'])):
    if df['MSSubClass'][i] not in sub_class:
        sub_class.append(df['MSSubClass'][i])
    if df['MSSubClass'][i] == 4 or df['MSSubClass'][i] == 0:
        print(df.iloc[i])
        
sub_class.pop(4)
sub_class.pop(0)

print(sub_class)

name = []
j = -1
for i in sub_class:
    j += 1
    name.append(str(i))
    name[j] = df.loc[df['MSSubClass'] == i].count()[0]


plt.pie(name, labels=sub_class, autopct='%.2f %%')
plt.title("MSSubClass")
plt.show()

plt.scatter(df['MSSubClass'], df['SalePrice'], alpha=0.5)
plt.show()

df = pd.get_dummies(df, dummy_na=False, columns=['MSSubClass'])
```
