```python
import joblib
import numpy as np
import pandas as pd

from mylib import preprocessing as prep
```

```python
model_filename = 'titanic_model.pkl'
str = 'test'
test_filename = str + '_formatted.csv'
orig_test_filename = str + '.csv'
# Load the model and test file
model = joblib.load(model_filename)
data = pd.read_csv(test_filename)
dtr = pd.read_csv('train_formatted.csv')
orig_data = pd.read_csv(orig_test_filename)
df = pd.DataFrame(data)
df_orig = pd.DataFrame(orig_data)
df_list = [df, dtr]
prep.quick_overview(df_list)
# Delete survived besauce it is a label.
# Delete Parch_9 because it does not exist in training data
col_name = ['Parch_9', 'Survived']
```

```python
prep.delete_unwanted_columns(df, col_name)
# Make predictions
predictions = model.predict(df)
# print(predictions)
```

```python
test_PassengerID = df_orig['PassengerId']
submission_df = {
    'PassengerId' : test_PassengerID,
    'Survived' : predictions
}
submission_df = pd.DataFrame(submission_df)
prep.view_on_html(submission_df)
```

```python
submission_df.to_csv('titanic_submission.csv', index=False)
```
